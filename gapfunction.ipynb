```julia


function draw(filename,Nx,Ny,zerostart) #use zerostart = true when ix = 0,1,2
    totalrows = countlines(filename)
    data = readlines(filename)
    mat_Δ = zeros(Complex{Float64},2*Nx,2*Ny)
    rx0 = div(Nx+1,2)
    ry0 = div(Ny+1,2)
    r = zeros(Float64,2)
    R = zeros(Float64,2)
    u1 = Float64[Nx,0]
    u2 = Float64[0, Ny]
    mφ = 2
    χ(r,R,n,m) = (-2*(π/2)*((R[1]*r[2]-R[2]*r[1])/(Nx*Ny)))-n*m*π#+n*π
        
    r = zeros(Float64,2)
    if zerostart
        for i in 1:totalrows
            line = data[i]
            #println(line)
            u = split(line)
            ix = parse(Int64,u[1])
            iy = parse(Int64,u[2])
            Δamp = parse(Float64,u[3])
            φ = parse(Float64,u[4])
            mat_Δ[ix+1,iy+1] = Δamp*exp(im*φ)
        end
    else
        count = 0
        for ix in 1:Nx
            for iy in 1:Ny
                count += 1

                line = data[count]
            #println(line)
                u = split(line)
            #ix = parse(Int64,u[1])
            #iy = parse(Int64,u[2])
                Δamp = parse(Float64,u[3])
                φ = parse(Float64,u[4])
                mat_Δ[ix,iy] = Δamp*exp(im*φ)
            end
            count += 1
        end
    end
    
    for ix in 1:Nx
        for iy in 1:Ny
            r[1] = ix - rx0
            r[2] = iy - ry0
            Δ = mat_Δ[ix,iy]
            
            #right
            m = 1
            n = 0
            R = u1
            φ = χ(r,R,n,m)
            
            mat_Δ[ix+Nx,iy] = Δ*exp(im*φ)
            #top
            m = 0
            n = 1
            R = u2
            φ = χ(r,R,n,m)
            
            mat_Δ[ix,iy+Ny] = Δ*exp(im*φ)
            #right top
            m = 1
            n = 1
            R = u1 + u2
            φ = χ(r,R,n,m)
            
            mat_Δ[ix+Nx,iy+Ny] = Δ*exp(im*φ)
                        
        end
    end
    
    fn = open("gap"*filename,"w")
    
    for ix in 1:2*Nx
        for iy in 1:2*Ny
            Δ = mat_Δ[ix,iy]
            φ = angle(Δ)
            φ += ifelse(φ > π,-2π,0)
            φ += ifelse(φ < -π,2π,0)
            println(fn,ix,"\t",iy,"\t",abs(Δ),"\t",φ)
        end
        println(fn," ")
    end
    close(fn)
    
end

Nx = 33
Ny = 33

draw("out_0.dat",Nx,Ny,true)
```
